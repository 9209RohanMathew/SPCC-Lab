```python
def charCheck(input_char):
     
    # CHECKING FOR ALPHABET
    if ((int(ord(input_char)) >= 65 and
        int(ord(input_char)) <= 90) or
        (int(ord(input_char)) >= 97 and
        int(ord(input_char)) <= 122)):
        return "Identifier"
     
    # CHECKING FOR DIGITS
    elif (int(ord(input_char)) >= 48 and
            int(ord(input_char)) <= 57):
        return "Digit"
    
    elif (int(ord(input_char))>=32 and int(ord(input_char))<=39) or (int(ord(input_char))==44 or int(ord(input_char)) ==46) or (int(ord(input_char))>=58 and int(ord(input_char))<=59) or ((int(ord(input_char)))==63):
        return "Special Character"
 
    # OTHERWISE Operator
    else:
        return "Operator"
```

```python
lst = []
exp = []
j = 100
while j > 0:
    print("1. Create Table  2. Search Table  3. Enter Symbol  4. Remove Symbol  5. View Table  6. Exit")
    choice = int(input("Enter your choice : "))
#     expr = str(input("Enter Expression : "))
    if choice == 1:
        expr = str(input("Enter Expression : "))
        for i in range(len(expr)):
            exp.append(expr[i])
        
    
    elif choice == 2:
        search = str(input("Enter Symbol to be searched : "))
        if search in exp:
            print("Symbol : ",search," "+"Address : ",id(search)," "+"Type : ",charCheck(search))
        else:
            print("Not Found")
    
    elif choice == 3:
        symb = str(input("Enter Symbol : "))
        if symb in exp:
            print("Duplicate Symbol")
        else:
            exp.append(symb)
        
    elif choice == 4:
        symb1 = str(input("Enter Symbol : "))
        if symb1 in exp: 
            exp.remove(symb1)
        else:
            print("Label not found")
            
    elif choice == 5:
        for i in exp:
            print("Symbol : ",i," "+"Address : ",id(i)," "+"Type : ",charCheck(i))
            
    elif choice == 6:
        j = -1
            
    else:
        print("Invalid Selection")
```
